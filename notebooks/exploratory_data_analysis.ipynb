````python
#Verificando se há duplicatas
trips_ohare.duplicated().sum()
````

````python
#Dropando os dados duplicados
trips_ohare.drop_duplicates(inplace=True)
trips_ohare.duplicated().sum()
````

````python
trips_ohare['weekday'] = pd.to_datetime(trips_ohare['start_ts']).dt.day_name()
trips_ohare[trips_ohare['weekday'] != 'Saturday'].head()
````

````python
#Agrupando os dias não chuvosos em uma variável nova
good_weather = trips_ohare[(trips_ohare['weather_conditions'] == 'Good') & (trips_ohare['weekday'] == 'Saturday')]
good_weather.head()
````

````python
#Agrupando o dias chuvosos em uma nova variável
bad_weather = trips_ohare[(trips_ohare['weather_conditions'] == 'Bad') & (trips_ohare['weekday'] == 'Saturday')]
bad_weather.head()
````

````python
#Criando uma coluna com o total de minutos para facilitar a visualização do tempo
good_weather['duration_minutes'] = (trips_ohare['duration_seconds'] / 60).round(1)
good_weather.head()
````

````python
#Criando uma coluna com o total de minutos para facilitar a visualização do tempo
bad_weather['duration_minutes'] = (trips_ohare['duration_seconds'] / 60).round(1)
bad_weather.head()
````

````python
bad_weather.shape, good_weather.shape
````

````python
#Fazendo o teste da hipótese

alpha = 0.5
results = st.ttest_ind(good_weather['duration_minutes'],
                         bad_weather['duration_minutes'])
print('valor-p: ', results.pvalue)

if (results.pvalue < alpha):
    print('Rejeitamos a hipótese nula')
else:
    print('Não podemos rejeitar a hipótese nula')
````

````python
#Com o teste realizado, obtivemos um valor-p extremamente baixo, por isso,
#podemos rejeitar a hipótese nula, o que significa que a chuva afeta na duração
#das corridas de taxi.
#Escolhi o valor alpha como 0.05 por não ser algo que necessite de uma enorme precisão,
#até porque o resultado não mudaria pelo o valor-p ser tão pequeno.
````

````python
print('Média dos dias chuvosos: ', bad_weather['duration_minutes'].mean().round(1))
print('Média dos dias não chuvosos: ', good_weather['duration_minutes'].mean().round(1))
````

````python
IQR = np.percentile(good_weather['duration_minutes'], 75) - np.percentile(good_weather['duration_minutes'], 25)
print('IQR: ', IQR)
````

````python
bin_width = 2 * IQR / (len(good_weather['duration_minutes']) ** (1/3))
print('Bin width: ', bin_width)
````

````python
fig = go.Figure()
fig.add_trace(go.Histogram(x=good_weather['duration_minutes'], nbinsx=26, name='Non rainy days'))
fig.add_trace(go.Histogram(x=bad_weather['duration_minutes'], nbinsx=26, name='Rainy days'))
fig.update_layout(title='Distribution of trip duration in minutes for rainy and non rainy days')
fig.update_layout(barmode='overlay')
fig.update_traces(opacity=0.75)
fig.show()
````
